If there is an object $x=\begin{bmatrix} x_1 & x_2 \end{bmatrix}$, we can think that the point corresponding to $x$ is on the 2-dimensional space. 


$
\begin{array}{ccc}
\text{student number} & \text{hrs. studying per day} & \text{GPA} \\
1 & 6 & 4.0 \\ 
2 & 10 & 4.3 \\
3 & 4 & 3.5 \\
4 & 1 & 2.7 \\
5 & 3 & 3.1 \\
6 & 7 & 3.7
\end{array}
$
![대체 텍스트](https://)




For example, we have a data set as the above. Then, we can show them  

```
import numpy as np
a = [6, 4.0]
b = [10, 4.3]
c = [4, 3.5]
d = [1, 2.7]
e = [3, 3.1]
f = [7, 3.7]
```

```
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10), dpi=100)
plt.plot(a[0], a[1], 'o', markersize=9, label='student 1')
plt.plot(b[0], b[1], 'o', markersize=9, label='student 2')
plt.plot(c[0], c[1], 'o', markersize=9, label='student 3')
plt.plot(d[0], d[1], 'o', markersize=9, label='student 4')
plt.plot(e[0], e[1], 'o', markersize=9, label='student 5')
plt.plot(f[0], f[1], 'o', markersize=9, label='student 6')
plt.xlabel('hrs.studying per day')
plt.ylabel('GPA')
plt.legend()
plt.grid('on')
```
